```python
from collections import deque

class Vertex:
    def __init__(self, name):
        self.name = name
        self.label = None
        self.adj = []

    def __repr__(self):
        return f"Vertex({self.name}, label={self.label})"


def AssignLabels(startVertex):

    next = deque()
    fel = set()             #Final Edge List (weights)
    unusedEdge = deque()
    state = {"gewMax": 0}   #Gloabl Edge Weight Maximum

    def printEdge(vA, vB, weight, src=""):
        print(f"EDGE {vA.name} - {vB.name} = {weight} {src}")

    def AddWeight(weight):
        fel.add(weight)
        if weight > state["gewMax"]:
            state["gewMax"] = weight

    def UpdateEdges(vertex):
        for adj in vertex.adj:
            if adj.label is not None:
                w = vertex.label + adj.label
                AddWeight(w)
                printEdge(vertex, adj, w, "[Derived Edge WEight]")

    def UpdateVertex(vertex, label, weight, src=""):
        vertex.label = label
        AddWeight(weight)
        print(f"Assigning label to {vertex.name}: label={label}, weight={weight} {src}")
        UpdateEdges(vertex)

    def AcceptibleLabel(examined, considered):
        for adj in examined.adj:
            if adj.label is not None:
                if considered == adj.label:
                    return False
                if considered + adj.label in fel:
                    return False
        return True

    startVertex.label = 1
    firstAdj = startVertex.adj[0]
    firstAdj.label = 1

    next.append(firstAdj)
    AddWeight(2)
    printEdge(startVertex, firstAdj, 2, "[initial]")


    #Main Loop#
    while next:
        Source = next.popleft()

        for Examinee in Source.adj:
            if Examinee.label is None:
                next.append(Examinee)
                usedQueue = False

                for weight in list(unusedEdge):
                    consideredWeight = weight
                    consideredLabel = consideredWeight - Source.label

                    if AcceptibleLabel(Examinee, consideredLabel):
                        UpdateVertex(Examinee, consideredLabel, consideredWeight, "[from unusedEdge]")
                        unusedEdge.remove(weight)
                        usedQueue = True
                        break

                if usedQueue:
                    continue

                consideredWeight = state["gewMax"] + 1

                while not AcceptibleLabel(Examinee, consideredWeight - Source.label):
                    consideredWeight += 1

                consideredLabel = consideredWeight - Source.label
                UpdateVertex(Examinee, consideredLabel, consideredWeight, "[new weight]")

                prev = state["gewMax"]
                if consideredWeight > prev + 1:
                    for w in range(prev + 1, consideredWeight):
                        unusedEdge.append(w)

    return fel, state["gewMax"], list(unusedEdge)
```

```python
#Temp vertices
A = Vertex('A')
B = Vertex('B')
C = Vertex('C')
D = Vertex('D')

A.adj = [B,C]
B.adj = [A,D]
C.adj = [A]
D.adj = [B]

AssignLabels(A)
```
